# Numpy is a Python package for manipulating lists and tables of numerical data. We can use it to do a lot of statistical calculations. We call the list or table of data a numpy array.

We often will take the data from our pandas DataFrame and put it in numpy arrays. Pandas DataFrames are great because we have the column names and other text data that makes it human readable. A DataFrame, while easy for a human to read, is not the ideal format for doing calculations. The numpy arrays are generally less human readable, but are in a format that enables the necessary computation.

We often start with our data in a Pandas DataFrame, but then want to convert it to a numpy array. The values attribute does this for us.

Let's convert the Fare column to a numpy array.

```python
import pandas as pd

# read_csv function takes a file in csv format and converts it to a Pandas DataFrame
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
```

```python
# The result is a 1-dimension array. why? Only one square bracket
df['Fare'].values
```

The values attribute of a Pandas Series give the data as a numpy array.

If we have a pandas DataFrame (instead of a Series as in the last part), we can still use the values attribute, but it returns a 2-dimensional numpy array.

```python
small_df = df[['Pclass', 'Fare', 'Age', 'Survived']]
arr = small_df.values
```

This is a 2-dimensional numpy array. You can tell because there’s two sets of brackets, and it expands both across the page and down

We use numpy shape attribute to determine the size of our numpy array

```python
print(arr.shape)
```

If we look at the shape, we get the number of rows and the number of columns

You can also use the shape attribute on a pandas DataFrame (df.shape)

```python
df.shape
```

```python
df.head()
print(arr)
```

```python
# select a single element from a numpy array with the following
arr[884, 3]
```

```python
# select a single row, for example, the whole row of the first passenger
arr[0]
```

```python
# To select a single column (in this case the Age column), we have to use some special syntax
arr[:,2]
```

# Masking

```python
# Often times you want to select all the rows that meet a certain criteria.
# select all the rows of children (passengers under the age of 18). 
mask = arr[:,2] < 18
```

A mask is a boolean array (True/False values) that tells us which values from the array we’re interested in

```python
arr[mask]
```

# Summing and Counting 

Let’s say we want to know how many of our passengers are children. We still have the same array definition and can take our mask or boolean values from the previous part.

Recall that True values are interpreted as 1 and False values are interpreted as 0. So we can just sum up the array and that’s equivalent to counting the number of true values.

```python
print(mask.sum())
```

Summing an array of boolean values gives the count of the number of True values
